```python
#random sampling nこに適当に分割
import random
import numpy as np

#random seed固定（仮）
random.seed(0)
#シャッフル
c_c_PT_ID_random = list(random.sample(c_c_PT_ID, len(c_c_PT_ID)))
#分割
pt_sample_splited = list(np.array_split(c_c_PT_ID_random, 1)) 
pt_sample_splited[0]
#並び替え＋ to list
sample_splited = []
for i in range(1):
    pt_sample_splited_one = list(pt_sample_splited[i])
    pt_sample_splited_one = sorted(pt_sample_splited_one)
    print(len(pt_sample_splited_one))
    sample_splited.append(pt_sample_splited_one)
sample_splited[0]
```

```python
#とりあえずsample_splited[0]で試します〜〜
r_c_c_PT_ID = sample_splited[0]

r_c_c_pt_no = []
for i in range(len(c_c_PT_ID)):
    one_c_c_PT_ID = c_c_PT_ID[i]
    if one_c_c_PT_ID in r_c_c_PT_ID:
        r_c_c_pt_no.append(i)
```

```python
#PT_ID表記
#death_ptとnot_death_ptを分ける
death_pt = []
not_death_pt = []
for i in range(len(c_c_PT_ID)):
    one_c_c_PT_ID = c_c_PT_ID[i]
    if one_c_c_PT_ID in c_c_death_PT_ID:
        death_pt.append(i)
    else:
        not_death_pt.append(i)
        
death_pt = set(r_c_c_pt_no) & set(death_pt)
not_death_pt = set(r_c_c_pt_no) & set(not_death_pt)

print(len(death_pt))
print(len(not_death_pt))
```

```python
list_r_c_c_pt_no = list(r_c_c_pt_no)
r_c_c_pt_no_1 = []
death_pt_no_1 = []
not_death_pt_no_1 = []
for i in range(len(list_r_c_c_pt_no)):
    r_c_c_pt_no_one = r_c_c_pt_no[i]
    r_c_c_pt_no_1.append(i)
    if r_c_c_pt_no_one in death_pt:
        death_pt_no_1.append(i)
    else:
        not_death_pt_no_1.append(i)
print(len(r_c_c_pt_no_1))
print(len(death_pt_no_1))
print(len(not_death_pt_no_1))
death_pt_no_1
```

```python
%%time
#ΔTをlast pointからの日数になるように計算
#かなり計算時間かかりそう〜〜
time_span = []

for i in r_c_c_pt_no:
    dfs_one = dfs[i]
    dfs_one = list(dfs_one['Δt'])
    dfs_steps_one = all_data_steps[i]
    dfs_one = dfs_one[0 : dfs_steps_one]
    dfs_one.reverse()
    
    time_sum = []
    for j in range(dfs_steps_one):
        dfs_one_sum = sum(dfs_one[0 : j])
        dfs_one_sum = dfs_one_sum * -1
        time_sum.append(dfs_one_sum)
    time_sum.reverse()
    
    #最初の3steps捨てる~!!!(改良)
    time_sum = time_sum[3 : ]
    time_span.append(time_sum)
    
time_span[0]
```
